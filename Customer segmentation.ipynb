```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
customer_data = pd.read_csv('Downloads/Mall_Customers.csv')
customer_data
```

```python
customer_data.shape
```

```python
customer_data.info()
```

```python
customer_data.isnull().sum()
```

```python
X = customer_data.iloc[:,[3,4]].values
print(X)
```

```python
#choosing the number of clusters
#finding wcss for diferent number of clusters

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
#creating the elbow graph to find which cluster has min value

sns.set()
plt.plot(range(1,11),wcss)
plt.title('The Elbow graph')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#optimum number of clusters is 5

#Training the model

kmeans = KMeans(n_clusters=5,init='k-means++',random_state=0)

#return a label for each cluster

Y = kmeans.fit_predict(X)

Y
```

```python
#visualizing all the clusters and centroid

plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green',label='Cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='red',label='Cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='yellow',label='Cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='violet',label='Cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='blue',label='Cluster 5')

#centroid

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='centroid')
plt.title('customers')
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.show()
```
